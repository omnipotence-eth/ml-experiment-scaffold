```python
from datasets import load_dataset

# Load your dataset
ds = load_dataset("openai/gsm8k", "main", split="train")
print(f"Rows: {len(ds)}")
print(f"Columns: {ds.column_names}")
ds[0]
```

```python
# Check answer distribution
import re

answers = []
for row in ds:
    match = re.search(r"#### ([-\d,]+)", row["answer"])
    if match:
        answers.append(float(match.group(1).replace(",", "")))

print(f"Parsed answers: {len(answers)} / {len(ds)}")
print(f"Min: {min(answers)}, Max: {max(answers)}, Mean: {sum(answers) / len(answers):.1f}")
```

```python
# Review random samples
import random

random.seed(42)
for i in random.sample(range(len(ds)), 3):
    print(f"=== Example {i} ===")
    print(f"Q: {ds[i]['question'][:200]}")
    print(f"A: {ds[i]['answer'][:200]}")
    print()
```
